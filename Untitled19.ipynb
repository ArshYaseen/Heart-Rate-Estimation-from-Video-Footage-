```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, find_peaks

# Constants
FS = 29.905  # Sampling frequency in Hz
CUTOFF = 1.0  # Cutoff frequency for low-pass filter in Hz
ORDER = 5  # Order of the Butterworth filter

def butter_lowpass_filter(data, cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y

def normalize_signal(signal):
    return (signal - np.min(signal)) / (np.max(signal) - np.min(signal))

def estimate_heart_rate(peaks, fs):
    if len(peaks) < 2:
        return None  # Not enough peaks to estimate heart rate
    peak_times = np.array(peaks) / fs
    intervals = np.diff(peak_times)
    average_interval = np.mean(intervals)  # in seconds
    heart_rate = 60 / average_interval
    return heart_rate

# Sample raw signal data (replace this with your actual signal data)
# This is just an example. Replace with your real signal data.
sample_data = np.array([
    120.5, 121.0, 121.5, 122.0, 122.5, 123.0, 123.5, 124.0, 124.5, 125.0,
    125.5, 126.0, 126.5, 127.0, 127.5, 128.0, 128.5, 129.0, 129.5, 130.0,
    130.5, 131.0, 131.5, 132.0, 132.5, 133.0, 133.5, 134.0, 134.5, 135.0
])

# Step 1: Signal Preprocessing
filtered_signal = butter_lowpass_filter(sample_data, CUTOFF, FS, ORDER)
normalized_signal = normalize_signal(filtered_signal)

# Step 2: Peak Detection
threshold = np.mean(normalized_signal) + 0.5 * np.std(normalized_signal)
min_distance = int(FS / 2)  # Minimum distance in samples (e.g., 2 seconds apart)

peaks, _ = find_peaks(normalized_signal, height=threshold, distance=min_distance)

# Step 3: Estimate Heart Rate
heart_rate = estimate_heart_rate(peaks, FS)

print(f"Detected Peaks: {peaks}")
print(f"Estimated Heart Rate: {heart_rate:.2f} BPM" if heart_rate is not None else "Not enough peaks detected.")

# Step 4: Visualization
plt.figure(figsize=(12, 6))
plt.plot(normalized_signal, label='Normalized Signal')
plt.plot(peaks, normalized_signal[peaks], 'rx', label='Detected Peaks')
plt.xlabel('Sample')
plt.ylabel('Amplitude')
plt.title('Signal with Detected Peaks')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, find_peaks

# Step 1: Create a Synthetic rPPG Signal (Simulated for demo)
def generate_synthetic_rppg(duration=10, fs=30, heart_rate=70):
    """
    Generates a synthetic rPPG-like signal for demonstration.
    :param duration: Duration of the signal in seconds.
    :param fs: Sampling frequency (frames per second).
    :param heart_rate: Simulated heart rate in BPM.
    :return: Synthetic rPPG signal.
    """
    t = np.linspace(0, duration, int(duration * fs))
    hr_freq = heart_rate / 60.0  # Convert BPM to Hz
    signal = np.sin(2 * np.pi * hr_freq * t) + 0.5 * np.random.randn(len(t))  # Add some noise
    return t, signal

# Step 2: Lowpass Butterworth Filter to Remove High-Frequency Noise
def butter_lowpass_filter(data, cutoff, fs, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y

# Step 3: Detect Peaks in the Signal
def detect_peaks(signal, fs):
    peaks, _ = find_peaks(signal, distance=fs//2)  # Distance is half a second to avoid noise
    return peaks

# Step 4: Calculate Heart Rate from Detected Peaks
def calculate_heart_rate(peaks, fs):
    if len(peaks) < 2:
        return None  # Not enough peaks to calculate heart rate
    peak_intervals = np.diff(peaks) / fs  # Time intervals between peaks
    avg_interval = np.mean(peak_intervals)
    heart_rate = 60.0 / avg_interval  # Convert to beats per minute
    return heart_rate

# Main function to run the pipeline
def main():
    # Parameters
    fs = 30  # Frames per second
    cutoff = 2.5  # Lowpass filter cutoff frequency in Hz
    duration = 10  # Duration of the synthetic signal in seconds
    true_heart_rate = 70  # Simulated heart rate in BPM

    # Step 1: Generate a synthetic rPPG-like signal
    t, synthetic_signal = generate_synthetic_rppg(duration=duration, fs=fs, heart_rate=true_heart_rate)

    # Step 2: Apply Lowpass Filter
    filtered_signal = butter_lowpass_filter(synthetic_signal, cutoff=cutoff, fs=fs)

    # Step 3: Detect Peaks
    peaks = detect_peaks(filtered_signal, fs)
    if len(peaks) < 2:
        print("Not enough peaks detected.")
        return

    # Step 4: Calculate Heart Rate
    estimated_heart_rate = calculate_heart_rate(peaks, fs)
    print(f"Estimated Heart Rate: {estimated_heart_rate:.2f} BPM")

    # Plot the signal and detected peaks
    plt.figure(figsize=(10, 6))
    plt.plot(t, filtered_signal, label='Filtered rPPG Signal')
    plt.plot(t[peaks], filtered_signal[peaks], 'ro', label='Detected Peaks')
    plt.title(f"Synthetic rPPG Signal - Estimated Heart Rate: {estimated_heart_rate:.2f} BPM")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
from scipy.signal import find_peaks, butter, lfilter
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import os

# Define the coordinates for the face ROI
ROI_COORDS = {
    'face': (0.30, 0.20, 0.70, 0.80)  # Adjusted to capture the whole face
}

def extract_roi(frame, roi_key):
    y1, x1, y2, x2 = ROI_COORDS[roi_key]
    height, width = frame.shape[:2]
    y1, x1, y2, x2 = int(y1 * height), int(x1 * width), int(y2 * height), int(x2 * width)
    return frame[y1:y2, x1:x2]

def calculate_mean_green(roi):
    return np.mean(roi[:, :, 1])

def draw_roi(frame, roi_key):
    y1, x1, y2, x2 = ROI_COORDS[roi_key]
    height, width = frame.shape[:2]
    y1, x1, y2, x2 = int(y1 * height), int(x1 * width), int(y2 * height), int(x2 * width)
    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
    return frame

def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    # Get video properties
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (int(cap.get(3)), int(cap.get(4))))

    signal = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Extract and draw ROI for the face
        face_roi = extract_roi(frame, 'face')
        frame = draw_roi(frame, 'face')
        
        # Calculate mean green channel value (rPPG signal) for the face region
        mean_green = calculate_mean_green(face_roi)
        signal.append(mean_green)
        
        # Write the frame with ROI to the output video
        out.write(frame)

    cap.release()
    out.release()
    return np.array(signal), fps

def butter_bandpass(lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

def plot_signal(signal, peaks, output_file='signal_plot.png'):
    plt.figure(figsize=(10, 6))
    plt.plot(signal, label='rPPG Signal')
    plt.plot(peaks, signal[peaks], 'rx', label='Detected Peaks')
    plt.title('rPPG Signal with Detected Peaks')
    plt.xlabel('Frame Index')
    plt.ylabel('Signal Amplitude')
    plt.legend()
    plt.grid(True)
    plt.savefig(output_file)
    plt.close()

def create_heart_rate_image(heart_rate, output_file='heart_rate.png'):
    width, height = 400, 200
    image = Image.new('RGB', (width, height), color=(255, 255, 255))
    draw = ImageDraw.Draw(image)
    
    try:
        font = ImageFont.truetype("arial.ttf", 30)
    except IOError:
        font = ImageFont.load_default()

    text = f"Estimated Heart Rate: {heart_rate:.2f} BPM"
    text_width, text_height = draw.textbbox((0, 0), text, font=font)[:2]  # Updated method
    
    position = ((width - text_width) / 2, (height - text_height) / 2)
    draw.text(position, text, fill=(0, 0, 0), font=font)
    
    image.save(output_file)

def estimate_heart_rate(video_path, output_path):
    rppg_signal, fps = process_video(video_path, output_path)
    if len(rppg_signal) == 0:
        print("No signal extracted.")
        return
    
    # Remove DC component
    rppg_signal = rppg_signal - np.mean(rppg_signal)

    # Apply bandpass filter
    lowcut = 0.7
    highcut = 3.0
    filtered_signal = butter_bandpass_filter(rppg_signal, lowcut, highcut, fps, order=4)

    # Detect peaks in the filtered signal
    peaks, _ = find_peaks(filtered_signal, distance=fps/2)

    # Calculate heart rate
    if len(peaks) < 2:
        print("Not enough peaks detected.")
        return
    
    peak_intervals = np.diff(peaks) / fps
    avg_interval = np.mean(peak_intervals)
    heart_rate = 60 / avg_interval
    
    print(f"Estimated Heart Rate: {heart_rate:.2f} BPM")

    # Save visualizations
    plot_signal(filtered_signal, peaks)
    create_heart_rate_image(heart_rate)

# Run the heart rate estimation
video_path = 'random1.mp4'  # Replace with your actual video path
output_path = 'output_with_rois.mp4'  # Path to save the video with ROIs
estimate_heart_rate(video_path, output_path)
```
